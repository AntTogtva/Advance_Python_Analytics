```python
import seaborn as sns 
```

```python
# basisc ploting a seaborn : 

tips=sns.load_dataset('tips')
tips.columns = tips.columns.str.strip()
```

```python
# avialble datasets in seaborn : 
sns.get_dataset_names()
```

```python
# seaborn made in the top of the matplotlib 

tips
```

```python
import matplotlib.pyplot as plt 

sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
import pandas as pd 
```

```python
sales = pd.read_csv('sales-data-sample.csv')
sales.columns = sales.columns.str.strip().str.lower()
sales.head(7)
```

```python
sns.scatterplot(x='profit',y='discount',data=sales)
plt.title('scatter plot of Profit - Discount')
plt.show()
```

```python
sns.lineplot(x='size',y='total_bill',data= tips )
plt.title('Size vs Total_bill')
plt.show()
```

```python
# categorical plots : 
# bar plot  

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Categorical day va total bill data ')
plt.show()
```

```python
# box plot 
sns.boxplot(x='day', y = 'total_bill',data=tips)
plt.title('Data Based on the day vs total_bill in box method ')
plt.show()
```

```python
# violin plot 

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('violinplot on day vs total_bill')
plt.show()
```

```python
# create an histogram using sns 

sns.histplot(tips['total_bill'],bins=10,kde=True) #kernel density is : get the curve that smooth the histogram : 
plt.title('histogram of total bill')
plt.show()
```

```python
# KDE plot 

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# pairplot 
sns.pairplot(tips)
```

```python
# heatmap 
import matplotlib.pyplot as plt
corr = tips[['total_bill','tip','size']].corr()
corr
sns.heatmap(corr,annot = True , cmap='coolwarm') # annot deplay the value in eacch cell and cmap set the color theme 

 # if want ot compute all the variables correlation ( with respect to all the dataset value then you can directly do that  but only work when the values is numeric  )

# to do that with using smoe inbuild function : 
numeric_tips = tips.select_dtypes(include="number")

# Compute correlation on numeric columns only
corr = numeric_tips.corr()

# Create the heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()

```

```python
from matplotlib import pyplot as plt 
```

```python
x = [num for num in range(1,6)]
y = [5,2,9,2,3]
print(x,y)
```

```python
plt.plot(x,y) # with respect to x and y axis 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("First Matplot-lib plot ")
```

```python
x = [num for num in range(1,6)]
y = [5,2,9,2,3]
print(x,y)
```

```python
# cutomized line plot 

plt.plot(x,y,color='blue',linestyle=':',marker='o',linewidth=3,markersize=9)
plt.grid(True) # creating a grid . 
```

```python
# creating a multiple plot a diffrent data 

x = [1,2,3,4,5,]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title('Plot-1')
plt.subplot(2,2,3)
plt.plot(x,y2,color='red')
plt.title('Plot-2')
plt.subplot(2,2,2)
plt.plot(y1,x,color='blue')
plt.title('Plot-3')
plt.subplot(2,2,4)
plt.plot(y2,x,color='pink')
plt.title('Plot - 4 ')
```

```python
# Bar ploting in matplotlib 

categories = ['A','B','C','D','E']
values = [5,7,3,8,9]

# create a bar plot 

plt.bar(categories,values,color='red')
plt.xlabel('Categories ')
plt.ylabel('values')
plt.title("the Bar graph vs categories vs values : ")
```

```python
sample_data = [ 1,2,2,3,3,6,6,11,11,12,12,3,3,4,4,9,10,15,9,8,3,13,13,4,2,2,2,5,5,5,5,5,7,7]

# create a histogram from this data 
plt.subplot(1,2,1)
plt.hist(sample_data,edgecolor='black')
plt.title('Plot-1')
plt.subplot(1,2,2)
plt.hist(sample_data,bins=7,color='orange',edgecolor='black')
plt.title('Plot-2')
```

```python
# Create a scatter plot 

x = [num for num in range(1,6)]
y = [5,2,9,2,3]
print(x,y)
plt.scatter(x,y,color='red',marker='x')
plt.grid()
```

```python
# pie chart 

labels = ['A','B','C','D']
sizes = [30,20,40,10]
colors =['yellow','red','blue','orange']
explode = (-0.2,0,0,0) # working on positive nagative way 

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%3.3f%%',shadow=True)
```

```python
# Real data set visulization : 

import pandas as pd 
```

```python
df=pd.read_csv('sales-data-sample.csv')
df.columns.to_list()
df.drop(columns=['OrderDate','City','CustomerName','PostalCode','Profit','ShipDate','DaystoShipActual','ShipStatus','DaystoShipScheduled','OrderProfitable','latitude','longitude'],axis=1 , inplace=True)
```

```python
df.head()
```

```python
profit_by_location =df.groupby('Region')['ProfitRatio'].sum()
print(profit_by_location)
```

```python
profit_by_location.plot(kind='bar',color='teal')
```
